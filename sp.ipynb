```python
import os
import pandas as pd
from bs4 import BeautifulSoup

# Define the folder path
folder_path = r"C:\Users\spraj\Downloads\JUPITER PROGRAMS\contechub\web scraping\scode"

# Get all .txt files in the folder
source_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.txt')]

all_data = []

# Loop through each file path
for file_path in source_files:
    if os.path.exists(file_path):
        with open(file_path, encoding='utf-8') as file:
            raw_html = file.read()
        soup = BeautifulSoup(raw_html, 'html.parser')

        Company_name = soup.find('h1', {'class': 'org-top-card-summary__title'}).text.strip() if soup.find('h1', {'class': 'org-top-card-summary__title'}) else ''
        Industry = soup.find('div', {'class': "org-top-card-summary-info-list__info-item"}).text.strip() if soup.find('div', {'class': "org-top-card-summary-info-list__info-item"}) else ''
        Company_size = soup.find('span', {'class': "t-normal t-black--light link-without-visited-state link-without-hover-state"}).text.strip() if soup.find('span', {'class': "t-normal t-black--light link-without-visited-state link-without-hover-state"}) else ''
        About_us = soup.find('div', {'id': 'about-us'}).text.strip() if soup.find('div', {'id': 'about-us'}) else ""
        specialities = soup.find('dd', {'class': "mb4 t-black--light text-body-medium", 'dir': "ltr"}).text.strip() if soup.find('dd', {'class': "mb4 t-black--light text-body-medium", 'dir': "ltr"}) else ''
        location = soup.find('p', {'class': "t-14 t-black--light t-normal break-words", 'dir': "ltr"}).text.strip() if soup.find('p', {'class': "t-14 t-black--light t-normal break-words", 'dir': "ltr"}) else ''
        Phone = soup.find('span', {'aria-hidden': "true", 'class': "link-without-visited-state", 'dir': "ltr"}).text.strip() if soup.find('span', {'aria-hidden': "true", 'class': "link-without-visited-state", 'dir': "ltr"}) else ''
        website = soup.find('span', {'class': "link-without-visited-state", 'dir': "ltr"}).text.strip() if soup.find('span', {'class': "link-without-visited-state", 'dir': "ltr"}) else ''
        Overview = soup.find('p', {'class': "break-words white-space-pre-wrap t-black--light text-body-medium"}).text.strip() if soup.find('p', {'class': "break-words white-space-pre-wrap t-black--light text-body-medium"}) else ''

        # Founded & Headquarters
        Founded = ''
        Headquarter = ''
        for dt_tag in soup.find_all('dt', {'class': 'mb1 text-heading-medium'}):
            dt_text = dt_tag.text.strip()
            if dt_text == "Founded":
                Founded = dt_tag.find_next('dd', {'class': 'mb4 t-black--light text-body-medium'}).text.strip()
            elif dt_text == "Headquarters":
                Headquarter = dt_tag.find_next('dd', {'class': 'mb4 t-black--light text-body-medium'}).text.strip()

        # Extract LinkedIn URL
        LinkedIn_url = ''
        a_tag = soup.find('a', {'class': 'ember-view active pv3 ph4 t-16 t-bold t-black--light org-page-navigation__item-anchor', 'href': True})
        if a_tag:
            href = a_tag['href']
            if href.startswith('/company/'):
                LinkedIn_url = f"https://www.linkedin.com{href}"

        # Check if any data was found
        data_found = any([Company_name, Industry, Founded, Company_size, About_us, specialities, location, Headquarter, website, Overview])

        # Extract file name from the path
        file_name = os.path.basename(file_path)

        data = {
            'File_Name': file_name,
            'Company_name': Company_name,
            'Phone': Phone,
            'Industry': Industry,
            'Founded': Founded,
            'Company_size': Company_size,
            'About_Us': About_us,
            'Specialities': specialities,
            'Location': location,
            'Headquarters': Headquarter,
            'Website': website,
            'Overview': Overview,
            'Data_Found': data_found,
            'LinkedIn_url': LinkedIn_url
        }

        all_data.append(data)
    else:
        print(f"File not found: {file_path}")

# Create a DataFrame
df = pd.DataFrame(all_data)

# Save DataFrame to Excel
output_path = r"C:\Users\spraj\Downloads\JUPITER PROGRAMS\contechub\web scraping\myoutput.xlsx"
df.to_excel(output_path, index=False)
print(f"Data successfully saved to {output_path}")
```
